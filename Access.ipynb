```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 1. SETUP & DATA LOADING
# Note: Government CSVs often use 'latin1' encoding due to French special characters.
try:
    access_df = pd.read_csv('acess_score.csv', encoding='latin1')
    print("Files loaded successfully.")
except FileNotFoundError:
    print("Error: Ensure the CSV files are in the same directory as this script.")

# 2. DATA CLEANING
# Convert 'Year-built' to numeric to allow for trend analysis
access_df['Year_Built_Clean'] = pd.to_numeric(access_df['Year-built_Annee-de-construction'], errors='coerce')

# Drop rows where critical data for correlation is missing
access_clean = access_df.dropna(subset=['Year_Built_Clean', 'Accessibility-score_Score-accessibilite'])

# 3. VISUALIZATION 1: Accessibility Score Distribution by Province
plt.figure(figsize=(12, 6))
province_accessibility = access_df.groupby('Province-territory_Province-territoire_eng')['Accessibility-score_Score-accessibilite'].mean().sort_values(ascending=False)

province_accessibility.plot(kind='bar', figsize=(12, 6), color='steelblue')
plt.title('Average Accessibility Score by Province', fontsize=14)
plt.ylabel('Average Accessibility Score')
plt.xlabel('Province')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.savefig('accessibility_by_province.png')

# 4. VISUALIZATION 2: The Modernization Trend (Age vs. Accessibility)
# We use a regression plot to see if newer buildings are significantly more accessible.
plt.figure(figsize=(10, 6))
sns.regplot(x='Year_Built_Clean', y='Accessibility-score_Score-accessibilite', data=access_clean, 
            scatter_kws={'alpha':0.4, 'color':'teal'}, line_kws={'color':'red'})

plt.title('The Modernization Gap: Building Age vs. Accessibility Score', fontsize=14)
plt.xlabel('Year Built')
plt.ylabel('Accessibility Score (0-100)')
plt.grid(True, linestyle='--', alpha=0.6)
plt.tight_layout()
plt.savefig('accessibility_trend.png')

# 5. KEY STATISTICS FOR THE INTERVIEW
# Calculate correlation coefficient (r)
correlation = access_clean[['Year_Built_Clean', 'Accessibility-score_Score-accessibilite']].corr().iloc[0, 1]

# Calculate statistics by province
print(f"\n--- Key Insights ---")
print(f"Correlation (r) between Age and Accessibility: {correlation:.2f}")
print(f"\nAverage Accessibility Score by Province:")
print(province_accessibility)
```
